neue Koordinaten

```python
z2=sp.Symbol('z2')
z2_p=sp.Symbol('z2_p')
alpha_1_p=sp.Symbol('alpha_1_p')
# Bestimmen x1_p und z2_p
x2_p=sp.Symbol("x2_p")
x1_p = z2+alpha_1 ##:
x3_bar=sp.Symbol("x3_bar")
# z2=x2-alpha_1
# print("z2:",z2)
# print("x1_p:",x1_p)
alpha_1_p = -k_1*x1_p ##:
x2_p=sys[1]
# z2_p=x2_p-alpha_1_p
# print(sp.expand(x2_p))
x2_p = sp.expand(x2_p) ##:
#Terme mit x3 durch x3_bar ersetzt
x2_p = 3.3677363097558*x2**2  - 4.52620710796917*x2 - 8.28920465708046+x3_bar ##:
# x2_p nur abhaengig von den variablen x1,z2,x3_bar
x2_p = x2_p.subs([(x2,z2+alpha_1)])##:
#x3_bar=- 0.142932856041132*x2*x3+0.00151658261343777*x3**2 + 0.0960502321843922*x3
# print("x2_p:",x2_p)
z2_p = x2_p-alpha_1_p ##:
# print("z2_p:",z2_p)
```

Schritt 3: Fiktiver Eingang $\overline{x}_3 = \alpha_2(x1,z2)$

Lyapunov-Funktion: $V_2 = \frac{1}{2}( x_1^2 + z_2^2)$

```python
V2 = 1/2*(x1-x1_r)**2+1/2*z2**2 ##:
# print(V2)
```

$\dot{V}_2 = x_1 \dot{x}_1 + z_2 \dot{z}_2$

```python
V2_p=sp.Symbol("V2_p")
alpha_2=sp.Symbol("alpha_2")
V2_p = (x1-x1_r)*x1_p+z2*z2_p ##:
# print((x1-x1_r)*x1_p)
x1_bar=sp.Symbol("x1_bar")#x1_bar= x1-x_r
phi_2=sp.Symbol("phi_2")
k_2=sp.Symbol("k_2")
# V2_p_temp = V2_p.subs([(x1-0.8,x1_bar)]) ##: 
#x1-0.8=x1_bar, der zu bestimmte fiktive Eingang alpha_2 ersetzt x3_bar.
V2_p = V2_p.subs([(x1-0.8,x1_bar),(x3_bar,alpha_2)]) ##:
V2_p = sp.expand(V2_p) ##:
# print(V2_p)
#Zwischenvariable zu bestimmen alpha_2 und phi_2. V2_p = -k1*x1^2+z2(alpha_2+phi_2)
V2_temp=V2_p-(- k_1*x1_bar**2) 
V2_temp=sp.expand(V2_temp/z2) 
# print(V2_temp)
phi_2 = V2_temp-alpha_2 ##:
# print("phi_2:",phi_2)
alpha_2 = -phi_2-k_2*z2 ##:
# print("alpha_2:",alpha_2)
#Prüfen Zeitableitung der Lyapunov-Funktion
V2_p_pruf = (alpha_2+phi_2)*z2-k_1*x1_bar**2 ##:
# print("V2_pruf:",V2_pruf)
```

$\dot{V}_2 = -k_1 \overline{x}^2_1 -k_2 z_2^2$

Schritt 4: Fehler $z_3:= \overline{x}_3-\alpha_{2}(x_1,z_2)$  für neue Koordinaten

```python
z3=sp.Symbol('z3')
z3_p=sp.Symbol('z3_p')
z2_p = z2_p.subs([(x3_bar,z3+alpha_2),(x1_bar,x1-x1_r)]) ##:
# print("z2_p",z2_p)
```

```python
x3_bar_p_temp=sp.Symbol('x3_bar_p_temp')
alpha_2_p=sp.Symbol('alpha_2_p')
x3_p = sys[2] ##：
# print(x3_p)
# Darstellung der x3_bar
x3_bar_p_temp = - 0.142932856041132*(x2_p*x3+x2*x3_p)+0.00151658261343777*2*x3_p*x3 + 0.0960502321843922*x3_p ##：
# print("x3_bar_p_temp",x3_bar_p_temp)
x3_bar_p_temp=x3_bar_p_temp.subs([(x3_bar,z3+alpha_2)])
# Abhängig vom x1,z2,z3,x3
x3_bar_p_temp = x3_bar_p_temp.subs([(x1_bar,x1-x1_r)])  ##:
# print("x3_bar_p_temp:",x3_bar_p_temp)
## Bechen Ableitung von z3
##Zeitableitung von phi_2(). lie-derivative of phi_2()
phi_2_lie = phi_2.subs([(x1_bar,x1-x1_r)]) ##:
# print(phi_2)
f_lie=sp.Matrix([[x1_p],[z2_p]])
xx_lie=sp.Matrix([[x1],[z2]])
#Zeitableitung von alpha2 -> alpha2_p
alpha_2_p = -k_2*z2_p-st.lie_deriv(phi_2_lie, f_lie,xx_lie ) ##:
# print(sp.expand(alpha_2_p))
#abhämgig von x1,z2,x3,u_pwm. Im folgenden schritt bestimmt man den Wert vom Eingang u_pwm
z3_p = x3_bar_p_temp-alpha_2_p ##:
# print("z3_p:",z3_p)
```

Schritt 5: Stabilisierung

$V_3 = \frac{1}{2}((x_1-x_{1r})^2+z_2^2+z_3^2$ 


$\dot{V}_3 = (x_1-x_{1r})\dot{x}_1+z_2\dot{z}_2+z_3\dot{z}_3$

```python
V3_p=sp.Symbol("V3_p")
k_3=sp.Symbol("k_3")
V3_p = (x1-x1_r)*x1_p+z2*z2_p+z3*z3_p ##:
# print(V3_p)
## V3_p=-k1*(x1-x1_r)**2-k2*z_2**2+z3*(phi_3+h(u_pwm,..))
V3_p_temp=V3_p-(-k_1*(x1-x1_r)**2-k_2*z2**2)
# print(sp.expand(V3_p_temp))
V3_p_temp = sp.expand(V3_p_temp/z3) ##:
# lass phi3(x1,z2,z3,x3)+h(u_pwm,...) = -k3*z3 nämlich phi3(x1,z2,z3,x3)+h(u_pwm,...)+k3*z3=0 -> V3_p= -k3*z3**2
V3_p_temp = V3_p_temp+k_3*z3 ##:
# print(sp.expand(V3_p_temp))
```

```python
##lösen u_pwm und substituieren
u_sol = sp.solve(V3_p_temp, u)[0] ##:
# print("u_sol with z3:",u_sol)
u_sol = u_sol.subs([(z3,x3_bar-alpha_2)]) ##:
# print("u_sol ",u_sol)
u_sol = u_sol.subs([(z2,x2-alpha_1)]) ##:
# print(u_sol)
u_sol = u_sol.subs([(x3_bar,- 0.142932856041132*x2*x3+0.00151658261343777*x3**2 + 0.0960502321843922*x3)]) ##:
u_sol = u_sol.subs([(x1_bar,x1-0.8)]) ##:
print("u_sol final",sp.expand(u_sol))
```
